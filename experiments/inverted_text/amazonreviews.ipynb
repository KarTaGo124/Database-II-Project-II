```python
import kagglehub

path = kagglehub.dataset_download("bittlingmayer/amazonreviews")
print("Path=", path)
```

```python
import os
print("Archivos encontrados:", os.listdir(path))
```

```python
import os
import bz2
import pandas as pd


def iter_texts_from_fasttext_bz2(fullpath):
    with bz2.open(fullpath, "rt", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            parts = line.split(" ", 1)
            if len(parts) < 2:
                continue
            yield parts[1]   # solo texto, sin __label__X


train_path = os.path.join(path, "train.ft.txt.bz2")
texts = []

for t in iter_texts_from_fasttext_bz2(train_path):
    texts.append(t)
    if len(texts) >= 64000:
        break

if len(texts) < 64000:
    test_path = os.path.join(path, "test.ft.txt.bz2")
    for t in iter_texts_from_fasttext_bz2(test_path):
        texts.append(t)
        if len(texts) >= 64000:
            break

print("Total textos cargados:", len(texts))

df = pd.DataFrame({
    "doc_id": range(len(texts)),
    "text": texts
})
```

```python
Ns = [1000, 2000, 4000, 8000, 16000, 32000, 64000]
OUTPUT_DIR = r"../../../data/datasets/amazon_reviews"
for N in Ns:
    out = os.path.join(OUTPUT_DIR, f"corpus_{N}.csv")
    df.iloc[:N].to_csv(out, index=False, encoding="utf-8")
    print(f"Archivo generado: {out} con {N} documentos")

print("Todos los CSV fueron generados correctamente.")
```
